In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are: "Brand
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time

def scrape_amazon(product_name):
    driver = webdriver.Chrome()
    driver.implicitly_wait(1) 

    products_data = []

    try:
        driver.get("https://www.amazon.in")
        search_box = driver.find_element(By.XPATH, "/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input")
        search_box.send_keys(product_name)
        search_box.send_keys(Keys.RETURN)

        for page in range(1, 2):
            time.sleep(3)

            products = driver.find_elements(By.CSS_SELECTOR, 'div[data-component-type="s-search-result"]')

            for product in products:
                try:
                    product_name = product.find_element(By.TAG_NAME, 'h2').text
                except:
                    product_name = '-'

                try:
                    price = product.find_element(By.CSS_SELECTOR, 'span.a-price-whole').text
                except:
                    price = '-'

                try:
                    product_url = product.find_element(By.TAG_NAME, 'a').get_attribute('href')
                except:
                    product_url = '-'

                if product_url != '-':
                    driver.execute_script("window.open(arguments[0]);", product_url)
                    driver.switch_to.window(driver.window_handles[-1])
                    time.sleep(3)
                    try:
                        brand_name = product.find_element(By.XPATH, '//div[@class="a-section a-spacing-small a-spacing-top-small"]/table/tbody/tr/td[2]').text
                    except:
                        brand_name = '-'

                    try:
                        return_exchange = driver.find_element(By.XPATH, '//span[@class="a-declarative"]/div[@class="a-section a-spacing-none icon-content"]/span[@class="a-size-small a-color-link a-text-normal"]').text
                    except:
                        return_exchange = '-'

                    try:
                        expected_delivery = driver.find_element(By.XPATH, '//div[@class="a-spacing-base"]/span/span').text
                    except:
                        expected_delivery = '-'

                    try:
                        availability = driver.find_element(By.ID, 'availability').text.strip()
                    except:
                        availability = '-'

                    driver.close()
                    driver.switch_to.window(driver.window_handles[0])

                    products_data.append({
                        "Brand Name": brand_name,
                        "Name of the Product": product_name,
                        "Price": price,
                        "Return/Exchange": return_exchange,
                        "Expected Delivery": expected_delivery,
                        "Availability": availability,
                        "Product URL": product_url
                    })

            try:
                next_page_button = driver.find_element(By.XPATH, '//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
                next_page_button.click()
            except:
                print(f"No more pages found. Scraped {page} page(s).")
                break

    except Exception as e:
        print(f"An error occurred: {e}")
    finally:
        driver.quit()

    df = pd.DataFrame(products_data)

    df.to_csv('amazon_products.csv', index=False)
    print("Data has been saved to amazon_products.csv")

if __name__ == "__main__":
    product_name = input("Enter the product name to search on Amazon.in: ")
    scrape_amazon(product_name)
```

```python
import os
import time
import requests
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

def fetch_image_urls(query, max_links_to_fetch, driver):
    search_url = f"https://www.google.com/search?hl=en&q={query}&tbm=isch"
    driver.get(search_url)

    image_urls = set()
    thumbnails = driver.find_elements(By.XPATH, '//div[@class="H8Rx8c"]')

    for img in thumbnails[:max_links_to_fetch]:
        try:
            img.click()
            time.sleep(1)
            images = driver.find_elements(By.XPATH, '//img[@class="sFlh5c pT0Scc"]')
            for image in images:
                src = image.get_attribute('src')
                if src and 'http' in src:
                    image_urls.add(src)
                    if len(image_urls) >= max_links_to_fetch:
                        break
        except Exception as e:
            print(f"Error: {e}")
            continue

    return list(image_urls)

def download_image(folder_path, url, file_name):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            with open(os.path.join(folder_path, file_name), 'wb') as f:
                f.write(response.content)
    except Exception as e:
        print(f"Could not download {url} - {e}")

def scrape_images():
    keywords = ['fruits', 'cars', 'Machine Learning', 'Guitar', 'Cakes']
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

    try:
        for keyword in keywords:
            print(f"Searching for {keyword}")
            folder_path = os.path.join(os.getcwd(), keyword)
            os.makedirs(folder_path, exist_ok=True)

            image_urls = fetch_image_urls(keyword, 10, driver)
            if not image_urls:
                print(f"No images found for {keyword}")
                continue

            for i, url in enumerate(image_urls):
                download_image(folder_path, url, f"{keyword}_{i+1}.jpg")
            print(f"Downloaded {len(image_urls)} images for keyword: {keyword}")

    finally:
        driver.quit()

if __name__ == "__main__":
    scrape_images()
```

Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps.

```python
driver.get("https://www.google.co.in/maps")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
import re

def get_coordinates(city_name):
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
    driver.implicitly_wait(10)
    
    try:
        driver.get("https://www.google.com/maps")
        
        search_box = driver.find_element(By.NAME, "q")
        search_box.send_keys(city_name)
        search_box.send_keys(Keys.RETURN)
        
        time.sleep(5)
        
        current_url = driver.current_url
        
        match = re.search(r'@([0-9.-]+),([0-9.-]+),', current_url)
        if match:
            latitude = match.group(1)
            longitude = match.group(2)
        else:
            latitude = None
            longitude = None
        
    except Exception as e:
        print(f"An error occurred: {e}")
        latitude = None
        longitude = None
    finally:
        driver.quit()
    
    return latitude, longitude

if __name__ == "__main__":
    city_name = input("Enter the city name to search on Google Maps: ")
    latitude, longitude = get_coordinates(city_name)
    if latitude and longitude:
        print(f"Coordinates of {city_name}: Latitude = {latitude}, Longitude = {longitude}")
    else:
        print(f"Could not find coordinates for {city_name}")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time

def scrape_hostelworld():
    driver = webdriver.Chrome()
    driver.implicitly_wait(5)

    hostelworld_data = []

    try:
        driver.get("https://www.hostelworld.com/")
        time.sleep(3)
        
        menu_button = driver.find_element(By.XPATH, '/html/body/div[3]/div/div[2]/main/header/nav/div[3]/div[2]/button')
        menu_button.click()
        
        accommodation_button = driver.find_element(By.XPATH, '/html/body/div[3]/div/div[2]/main/header/nav/div[3]/div[2]/div[1]/div[2]/div/ul/li[1]/button')
        accommodation_button.click()
        
        hostels_button = driver.find_element(By.XPATH, '/html/body/div[3]/div/div[2]/main/header/nav/div[3]/div[2]/div[1]/div[2]/div/ul/li[1]/div/div/ul/li[1]/a')
        hostels_button.click()
        
        london_hostels = driver.find_element(By.XPATH, '/html/body/div[3]/div/div[2]/main/section[1]/div[2]/div[2]/a')
        london_hostels.click()

        for page in range(1, 5):
            time.sleep(3)

            listings = driver.find_elements(By.XPATH, '//a[@class="property-card-container property-listing-card horizontal"]')

            for listing in listings:
                try:
                    hostel_name = listing.find_element(By.XPATH, './/div[@class="property-info-container"]/div[@class="property-info-header"]/div[@class="property-name"]/span').text
                except:
                    hostel_name = '-'

                try:
                    hostel_rating = listing.find_element(By.XPATH, './/div[@class="property-rating"]/div/div[@class="rating-score"]/span[1]').text
                except:
                    hostel_rating = '-'

                try:
                    total_reviews = listing.find_element(By.XPATH, '//div[@class="property-rating"]/div[@class="rating theme-null"]/div[@class="review"]').text
                except:
                    total_reviews = '-'

                try:
                    overall_review = listing.find_element(By.XPATH, './/div[@class="property-rating"]/div/div[@class="rating-score"]/span[2]').text
                except:
                    overall_review = '-'

                try:
                    distance = listing.find_element(By.XPATH, './/div[@class="property-distance"]/span[2]').text
                except:
                    distance = '-'
                
                try:
                    privates = listing.find_element(By.XPATH, './/div[@class="property-accommodation-prices"]/div/div[@class="accommodation-price single-row"]/strong').text
                except:
                    privates = '-'
                
                try:
                    dorms = listing.find_element(By.XPATH, './/div[@class="property-accommodation-prices"]/div[2]/div[@class="accommodation-price single-row"]/strong').text
                except:
                    dorms = '-'

                try:
                    description = listing.find_element(By.XPATH, './/div[@class="property-description"]/span').text
                except:
                    description = '-'
                
                try:
                    listing_url = listing.get_attribute('href')
                except:
                    listing_url = '-'

                facilities_dict = {}
                if listing_url != '-':
                    driver.execute_script("window.open(arguments[0]);", listing_url)
                    driver.switch_to.window(driver.window_handles[-1])
                    time.sleep(3)
                    try:
                        facilities_section = driver.find_element(By.CLASS_NAME, 'property-facilities-section')
                        categories = facilities_section.find_elements(By.TAG_NAME, 'li')
                        for category in categories:
                            category_name = category.find_element(By.TAG_NAME, 'span').text.strip()
                            facilities = category.find_elements(By.TAG_NAME, 'li')
                            facilities_list = [facility.text.strip() for facility in facilities]
                            facilities_dict[category_name] = facilities_list
                    except:
                        pass
                    
                    driver.close()
                    driver.switch_to.window(driver.window_handles[0])

                hostelworld_data.append({
                    "Hostel Name": hostel_name,
                    "Rating": hostel_rating,
                    "Total Reviews": total_reviews,
                    "Overall Review": overall_review,
                    "Distance": distance,
                    "Privates Price": privates,
                    "Dorms Price": dorms,
                    "Description": description,
                    "Listing URL": listing_url,
                    "Facilities": facilities_dict
                })

            try:
                next_page_button = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/div[3]/main/div[2]/section[1]/div[3]/div/a')
                next_page_button.click()
                page += 1
            except:
                print(f"No more pages found. Scraped {page} page(s).")
                break

    except Exception as e:
        print(f"An error occurred: {e}")
    finally:
        driver.quit()

    df = pd.DataFrame(hostelworld_data)
    print(df)

if __name__ == "__main__":
    scrape_hostelworld()
```
