1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException, WebDriverException
import time
import pandas as pd

driver = webdriver.Chrome()

url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"

videos_data = []

try:
    driver.get(url)
    
    time.sleep(3)
    
    table = driver.find_element(By.XPATH, '//*[@id="mw-content-text"]/div[1]/table[1]')
    
    rank_counter = 1

    for row in table.find_elements(By.TAG_NAME, 'tr')[1:]: 
        try:
            cells = row.find_elements(By.TAG_NAME, 'td')
            if len(cells) == 5:
                name = cells[0].text.split('[')[0].strip().replace('"', '').strip()
                artist = cells[1].text
                views = cells[2].text
                upload_date = cells[3].text
                views = views.split('[')[0].strip()
                
                videos_data.append({
                    "Rank": rank_counter,
                    "Name": name,
                    "Artist": artist,
                    "Upload Date": upload_date,
                    "Views (Billions)": views
                })
                
                rank_counter += 1
        except NoSuchElementException as e:
            print(f"Element not found: {e}")
        except IndexError as e:
            print(f"Index error: {e}")

except WebDriverException as e:
    print(f"WebDriverException: {e}")

finally:
    driver.quit()

df = pd.DataFrame(videos_data)

df
```

3. Scrape the details of State-wise GDP of India from statisticstime.com.
Url = http://statisticstimes.com/
You have to find following details: A) Rank
B) State
C) GSDP(18-19)- at current prices
D) GSDP(19-20)- at current prices
E) Share(18-19)
F) GDP($ billion)
Note: - From statisticstimes home page you have to reach to economy page through code. 

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd

driver = webdriver.Chrome()

driver.get("https://statisticstimes.com/economy/india/indian-states-gdp.php") 

rows = driver.find_elements(By.XPATH, "//table[@id='table_id']/tbody/tr")

rank = []
state = []
gsdp_18_19 = []
gsdp_19_20 = []
share_18_19 = []
gdp_billion = []

for row in rows:
    cells = row.find_elements(By.TAG_NAME, "td")
    
    if len(cells) > 1:
        rank.append(cells[0].text)
        state.append(cells[1].text)
        gsdp_18_19.append(cells[2].text)
        gsdp_19_20.append(cells[3].text)
        share_18_19.append(cells[5].text)
        gdp_billion.append(cells[6].text)


driver.quit()

data = {
    "Rank": rank,
    "State": state,
    "GSDP(18-19) - at current prices": gsdp_18_19,
    "GSDP(19-20) - at current prices": gsdp_19_20,
    "Share(18-19)": share_18_19,
    "GDP($ billion)": gdp_billion
}

df = pd.DataFrame(data)

print(df)
```

4. Scrape the details of trending repositories on Github.com.
Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used
Note: - From the home page you have to click on the trending option from Explore menu through code. 

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

driver = webdriver.Chrome()

driver.get("https://github.com/")

opensource_menu = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.XPATH, "/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[4]/button"))
)
opensource_menu.click()

trending_link = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.XPATH, "/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[4]/div/div[3]/ul/li[2]/a"))
)
trending_link.click()

WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.XPATH, "//article[@class='Box-row']"))
)

repo_titles = []
repo_descriptions = []
contributors_counts = []
languages_used = []

repos = driver.find_elements(By.XPATH, "//article[@class='Box-row']")

for repo in repos:
    title = repo.find_element(By.XPATH, './/a[@class="Link"]').text
    repo_titles.append(title)
    
    try:
        description = repo.find_element(By.XPATH, './/p[@class="col-9 color-fg-muted my-1 pr-4"]').text
    except:
        description = "No description"
    repo_descriptions.append(description)
    
    try:
        contributors = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.XPATH, "//span[@data-view-component='true']//a[@class='d-inline-block']")))

        num_contributors = len(contributors)
    except:
        contributors = "0"
    contributors_counts.append(num_contributors)
    
    try:
        language = repo.find_element(By.XPATH, ".//span[@itemprop='programmingLanguage']").text
    except:
        language = "No language specified"
    languages_used.append(language)

driver.quit()

data = {
    "Repository Title": repo_titles,
    "Repository Description": repo_descriptions,
    "Contributors Count": contributors_counts,
    "Language Used": languages_used
}

df = pd.DataFrame(data)

print(df)
```

5. Scrape the details of top 100 songs on billiboard.com. Url = https:/www.billboard.com/ You have to find the
following details:
A) Song name
B) Artist name
C) Last week rank
D) Peak rank
E) Weeks on board
 Note: - From the home page you have to click on the charts option then hot 100-page link through code. 

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

driver = webdriver.Chrome()

url = "https://www.billboard.com/"

driver.get(url)

charts=driver.find_element(By.XPATH,'/html/body/div[3]/header/div/div[2]/div/div/div[2]/div[2]/div/div/nav/ul/li[1]/a')

charts.click()

top100=driver.find_element(By.XPATH,'/html/body/div[3]/main/div[2]/div[1]/div[1]/div/div/div[1]/div/div[2]/span/a')

top100.click()

ranks=[]
song_names = []
artist_names = []
last_week_ranks = []
peak_ranks = []
weeks_on_board = []

chart_rows = driver.find_elements(By.XPATH, './/div[@class="o-chart-results-list-row-container"]')

for row in chart_rows:
    song_name = row.find_element(By.XPATH, './/h3[@id="title-of-a-story"]').text
    artist_name = row.find_element(By.XPATH, './/li[@class="lrv-u-width-100p"]/ul/li[1]/span[1]').text
    last_week_rank = row.find_element(By.XPATH, './/li[@class="lrv-u-width-100p"]/ul/li[4]/span').text
    peak_rank = row.find_element(By.XPATH, './/li[@class="lrv-u-width-100p"]/ul/li[5]/span').text
    weeks_on_chart = row.find_element(By.XPATH, './/li[@class="lrv-u-width-100p"]/ul/li[6]/span').text

    song_names.append(song_name)
    artist_names.append(artist_name)
    last_week_ranks.append(last_week_rank)
    peak_ranks.append(peak_rank)
    weeks_on_board.append(weeks_on_chart)

df = pd.DataFrame({
    "Song Name": song_names,
    "Artist Name": artist_names,
    "Last Week Rank": last_week_ranks,
    "Peak Rank": peak_ranks,
    "Weeks on Board": weeks_on_board
})

print(df)

driver.quit()
```

6. Scrape the details of Highest selling novels.
A) Book name
B) Author name
C) Volumes sold
D) Publisher
E) Genre
Url - https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException

url = "https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare"
driver = webdriver.Chrome()

try:
    driver.get(url)
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//table[@class='in-article sortable']")))

    rows = driver.find_elements(By.XPATH, "//table[@class='in-article sortable']/tbody/tr")
    ranks=[]
    book_names = []
    author_names = []
    volumes_sold = []
    publishers = []
    genres = []

    for row in rows:
        try:
            columns = row.find_elements(By.XPATH, "./td")
            if len(columns) >= 5:
                rank=columns[0].text.strip()
                book_name = columns[1].text.strip()
                author_name = columns[2].text.strip()
                volume = columns[3].text.strip()
                publisher = columns[4].text.strip()
                genre = columns[5].text.strip()
                
                ranks.append(rank)
                book_names.append(book_name)
                author_names.append(author_name)
                volumes_sold.append(volume)
                publishers.append(publisher)
                genres.append(genre)
        
        except NoSuchElementException as e:
            print(f"Element not found: {e}")
        
        except Exception as e:
            print(f"An error occurred: {e}")

    df = pd.DataFrame({
        'Rank':ranks,
        'Book Name': book_names,
        'Author Name': author_names,
        'Volumes Sold': volumes_sold,
        'Publisher': publishers,
        'Genre': genres
    })

    print(df)

except TimeoutException:
    print("Timeout occurred waiting for the table to load.")

except Exception as e:
    print(f"Unexpected error: {e}")

finally:
    driver.quit()
```

7. Scrape the details most watched tv series of all time from imdb.com.
Url = https://www.imdb.com/list/ls095964455/ You have to find the following details:
A) Name
B) Year span
C) Genre
D) Run time
E) Ratings
F) Votes

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import re

def extract_rating(rating_str):
    match = re.search(r'([^\(\)]+)', rating_str)
    if match:
        return match.group(1).strip()
    return None

driver = webdriver.Chrome()

url = "https://www.imdb.com/chart/tvmeter/?ref_=nv_tvv_mptv"
driver.get(url)

WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.CSS_SELECTOR, ".ipc-metadata-list-summary-item"))
)

series_names = []
year_spans = []
genres = []
runtimes = []
ratings = []
votes = []

series_elements = driver.find_elements(By.XPATH, './/li[@class="ipc-metadata-list-summary-item sc-10233bc-0 iherUv cli-parent"]')

for series in series_elements:
    series_name = None
    year_span = None
    genre = None
    runtime = None
    rating_a = None
    votes_count = None

    try:
        series_name_element = series.find_element(By.XPATH, './/div[@class="ipc-title ipc-title--base ipc-title--title ipc-title-link-no-icon ipc-title--on-textPrimary sc-b189961a-9 iALATN cli-title"]/a')
        series_name = series_name_element.text.strip() if series_name_element else None

        year_span_element = series.find_element(By.XPATH, './/div[@class="sc-b189961a-7 feoqjK cli-title-metadata"]/span[1]')
        year_span = year_span_element.text.strip() if year_span_element else None

        genre_element = series.find_element(By.XPATH, './/div[@class="sc-b189961a-0 hBZnfJ cli-children"]/div[@class="sc-b189961a-7 feoqjK cli-title-metadata"]/span[3]')
        genre = genre_element.text.strip() if genre_element else None

        runtime_element = series.find_element(By.XPATH, './/div[@class="sc-b189961a-7 feoqjK cli-title-metadata"]/span[2]')
        runtime = runtime_element.text.strip() if runtime_element else None

        rating_element = series.find_element(By.XPATH, './/div[@class="sc-e2dbc1a3-0 ajrIH sc-b189961a-2 fkPBP cli-ratings-container"]/span')
        rating = rating_element.text.strip() if rating_element else None
        rating_a = extract_rating(rating) if rating else None

        votes_count_element = series.find_element(By.XPATH, './/span[@class="ipc-rating-star--voteCount"]')
        votes_count = votes_count_element.text.strip().replace("(", "").replace(")", "") if votes_count_element else None

    except Exception as e:
        print(f"Error extracting series data: {str(e)}")

    series_names.append(series_name)
    year_spans.append(year_span)
    genres.append(genre)
    runtimes.append(runtime)
    ratings.append(rating_a)
    votes.append(votes_count)

driver.quit()

data = {
    "Name": series_names,
    "Year Span": year_spans,
    "Genre": genres,
    "Run Time": runtimes,
    "Rating": ratings,
    "Votes": votes
}

df = pd.DataFrame(data)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

def extract_content(element):
    return element.text.strip() if element else None

def scrape_uci_datasets(url):
    driver = webdriver.Chrome()
    driver.get(url)

    try:
        popular_dataset=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/section[1]/div[2]/a')

        popular_dataset.click()
        dataset_elements = driver.find_elements(By.CSS_SELECTOR, 'div.rounded-box.bg-base-100')

        dataset_details = []
        expand_all=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[1]/div/label[2]')
        expand_all.click()
        for dataset_element in dataset_elements:
            details = {}

            name_elem = dataset_element.find_element(By.CSS_SELECTOR, 'h2.truncate.text-primary a.link-hover.link.text-xl.font-semibold')
            details['Dataset Name'] = extract_content(name_elem)

            desc_elem = dataset_element.find_element(By.CSS_SELECTOR, 'p.truncate')
            details['Description'] = extract_content(desc_elem)

            details_table = dataset_element.find_element(By.CSS_SELECTOR, 'div.my-2.hidden.gap-4.md\:grid.grid-cols-12')
            if details_table:
                rows = details_table.find_elements(By.CSS_SELECTOR, 'div.col-span-3.flex.items-center.gap-2')
                if len(rows) >= 4:
                    details['Task'] = extract_content(rows[0]).replace('Task', '').strip()
                    details['Data Type'] = extract_content(rows[1]).replace('Feature Type', '').strip()
                    details['No of Instances'] = extract_content(rows[2]).replace('150 Instances', '').strip()
                    details['No of Attributes'] = extract_content(rows[3]).replace('4 Features', '').strip()

            dataset_details.append(details)

        return dataset_details

    finally:
        driver.quit()

url = 'https://archive.ics.uci.edu/'

dataset_details = scrape_uci_datasets(url)

df = pd.DataFrame(dataset_details)

print(df)
```
