```python
start = 0
end = 3

brand_count = 0
description_count = 0
price_count = 0

brand_100 = []
product_desc_100 = []
price_100 = []

for page in range(start, end):
    if brand_count >= 100 and description_count >= 100 and price_count >= 100:
        break

    brands = driver.find_elements(By.XPATH, '//div[@class="syl9yP"]')
    for b in brands:
        if brand_count < 100:
            brand = b.text
            brand_100.append(brand)
            brand_count += 1
        else:
            break

    product_desc = driver.find_elements(By.XPATH, '//div[@class="hCKiGj"]//a[@class="WKTcLC" or @class="WKTcLC BwBZTg"]')
    for review in product_desc:
        if description_count < 100:
            description = review.text
            product_desc_100.append(description)
            description_count += 1
        else:
            break

    prices = driver.find_elements(By.XPATH, '//div[@class="Nx9bqj"]')
    for pr in prices:
        if price_count < 100:
            price = pr.text
            price_100.append(price)
            price_count += 1
        else:
            break

    next_button = driver.find_element(By.XPATH, '/html/body/div/div/div[3]/div[1]/div[2]/div[12]/div/div/nav/a[11]')
    next_button.click()
    time.sleep(3)
```

```python
i7_button=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div[2]/div/div[3]/span/div[1]/div/div/div[6]/ul[19]/span/span[9]/li/span/a/span')

i7_button.click()
```

```python
start = 0
end = 2

title_count = 0
rating_count = 0
price_count = 0

title_10 = []
rating_10 = []
price_10 = []

for page in range(start, end):
    if title_count >= 10 and rating_count >= 10 and price_count >= 10:
        break

    titles = driver.find_elements(By.XPATH, '//h2[@class="a-size-mini a-spacing-none a-color-base s-line-clamp-2"]')
    for t in titles:
        if title_count < 10:
            title = t.text
            title_10.append(title)
            title_count += 1
        else:
            break

    ratings = driver.find_elements(By.XPATH, '//div[@class="a-row a-size-small"]//span[@aria-label][1]')
    for r in ratings:
        if rating_count < 10:
            rating = r.get_attribute("aria-label")
            rating_value = rating.split(' out')[0]
            rating_10.append(rating_value)
            rating_count += 1
        else:
            break

    prices = driver.find_elements(By.XPATH, '//span[@class="a-price"]')
    for pr in prices:
        if price_count < 10:
            price = pr.text
            price_10.append(price)
            price_count += 1
        else:
            break
```

Q6: Write a python program to scrape data for Top 1000 Quotes of All Time.
The above task will be done in following steps:
1. First get the webpagehttps://www.azquotes.com/
2. Click on Top Quote
3. Than scrap a) Quote b) Author c) Type Of Quotes

```python
driver.get("https://www.azquotes.com/")
```

```python
top_quotes_button=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')

top_quotes_button.click()
```

```python
start = 0
end = 10

quote_count = 0
author_count = 0
typeq_count = 0

quote_1000 = []
author_1000 = []
typeq_1000 = []

for page in range(start, end):
    if quote_count >= 1000 and author_count >= 1000 and typeq_count >= 1000:
        break

    quotes = driver.find_elements(By.XPATH, '//a[@class="title"]')
    for q in quotes:
        if quote_count < 1000:
            quote = q.text
            quote_1000.append(quote)
            quote_count += 1
        else:
            break

    authors = driver.find_elements(By.XPATH, '//div[@class="author"]/a')
    for a in authors:
        if author_count < 1000:
            author = a.text
            author_1000.append(author)
            author_count += 1
        else:
            break

    type_quotes = driver.find_elements(By.XPATH, '//div[@class="tags"]')
    for tq in type_quotes:
        if typeq_count < 1000:
            type_quote = tq.text
            typeq_1000.append(type_quote)
            typeq_count += 1
        else:
            break

    next_button = driver.find_element(By.XPATH, '/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[3]/li[12]/a')
    next_button.click()
    time.sleep(5)
```

```python
quote_df=pd.DataFrame({'Quote':quote_1000,'Author':author_1000,'Type of Quote':typeq_1000})

quote_df
```

Q7: Write a python program to display list of respected former Prime Ministers of India (i.e. Name, Born-Dead, Term of office, Remarks) from https://www.jagranjosh.com/general-knowledge/list-of-all-prime-ministers-of-india-1473165149-1
scrap the mentioned data and make the DataFrame

```python
driver.get("https://www.jagranjosh.com/general-knowledge/list-of-all-prime-ministers-of-india-1473165149-1")
```

```python
rows=len(driver.find_elements(By.XPATH,"/html/body/div[1]/main/div[1]/div[1]/article/div[3]/div[7]/div/table/tbody/tr"))
cols=len(driver.find_elements(By.XPATH,"/html/body/div[1]/main/div[1]/div[1]/article/div[3]/div[7]/div/table/tbody/tr[2]/td"))


print(rows)
print(cols)
```

```python
columns_data = {f'Column_{i}': [] for i in range(1, cols + 1)}
columns_data['Term of Office'] = []


for r in range(2, rows + 1):
    for c in range(1, cols + 1):
        value = driver.find_element(By.XPATH, f"/html/body/div[1]/main/div[1]/div[1]/article/div[3]/div[7]/div/table/tbody/tr[{r}]/td[{c}]").text
        if c == 4:
            if '\n' in value:
                period, duration = value.split('\n')
                term_of_office = f"{period.strip()} ({duration.strip()})"
            else:
                term_of_office = value.strip()
            columns_data['Term of Office'].append(term_of_office)
        else:
            columns_data[f'Column_{c}'].append(value)

del columns_data['Column_4']


PM_df = pd.DataFrame(columns_data)


PM_df
```

```python
PM_df.drop('Column_1',inplace=True,axis=1)
```

```python

column_names={
    'Column_2':'Name',
    'Column_3':'Born_Dead',
    'Column_5':'Remarks'
}
PM_df.rename(columns=column_names, inplace=True)

PM_df = PM_df[['Name', 'Born_Dead', 'Term of Office', 'Remarks']]
```

```python
start = 0
end = 1

name_count = 0
price_count = 0

name_50 = []
price_50 = []

for page in range(start, end):
    if name_count >= 50 and price_count >= 50:
        break

    names = driver.find_elements(By.CLASS_NAME, 'subheader')
    for n in names:
        if name_count < 50:
            name = n.text
            name_50.append(name)
            name_count += 1
        else:
            break

    prices = driver.find_elements(By.XPATH, '//p/strong')
    for pr in prices:
        if price_count < 50:
            price = pr.text
            price_50.append(price)
            price_count += 1
        else:
            break
```
