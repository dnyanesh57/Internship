4) Write a python program to scrape first 10 product details which include product name , price , Image URL from
https://www.bewakoof.com/bestseller?sort=popular .

```python
html_text_b = requests.get('https://www.bewakoof.com/bestseller?sort=popular').text

soup_b = BeautifulSoup(html_text_b, "html.parser")

product_cards=soup_b.find_all("div",class_="plp-product-card")

data = []

def extract_price(price_str):
    match = re.search(r'(₹\d+(?:,\d{3})*(?:\.\d+)?)(?:[^\d.]|$)', price_str)
    return match.group(1) if match else None

for product_card in product_cards:
    title_element = product_card.find('h2')
    act_price_element = product_card.find("div" ,class_="actualPriceText clr-shade5")
    disc_price_element = product_card.find("div" ,class_="discountedPriceText clr-p-black false")
    loyalty_price_element= product_card.find("div" ,class_="loyaltyPriceBox")
    img_url_element= product_card.find("img",class_="productImgTag")['src']

    
    title = title_element.get_text(strip=True) if title_element else None
    act_price = act_price_element.get_text(strip=True) if act_price_element else None
    disc_price = disc_price_element.get_text(strip=True) if disc_price_element else None
    loyalty_price_str = loyalty_price_element.get_text(strip=True) if loyalty_price_element else None
    loyalty_price = extract_price(loyalty_price_str) if loyalty_price_str else None
    img_url=img_url_element if img_url_element else None
    
    data.append({
        'Product Name': title,
        'Actual Price': act_price,
        'Discounted Price': disc_price,
        'Price for TriBe Members':loyalty_price,
        'Image URL':img_url
    })


bewakoof_df = pd.DataFrame(data)

bewakoof_df.head(10)
```

5) Please visit https://www.cnbc.com/world/?region=world and scrapa)
headings
b) date
c) News link

```python
def extract_date(url):
    date_pattern = r'/(\d{4})/(\d{2})/(\d{2})/'
    match = re.search(date_pattern, url)
    if match:
        year, month, day = match.groups()
        return f"{year}-{month}-{day}"
    else:
        return None

html_text_cnbc = requests.get('https://www.cnbc.com/world/?region=world').text
soup_cnbc = BeautifulSoup(html_text_cnbc, "html.parser")

data = []

for i in range(21):
    if i not in (8, 9):
        div_id = f"HomePageInternational-riverPlus-11-{i}"
        news_card = soup_cnbc.find("div", id=div_id)
        
        if news_card:
            heading_element = news_card.find('a', href=lambda href: href and href.startswith('https://') and '/pro/' not in href)
            if heading_element:
                heading = heading_element.get_text(strip=True)
                news_link_element = heading_element['href']
                news_link = news_link_element
                news_date = extract_date(news_link)
                
                data.append({
                    'Heading': heading,
                    'Date': news_date,
                    'News Link': news_link
                })
            else:
                headline_element = news_card.find('div', class_='RiverHeadline-headline')
                if headline_element:
                    heading = headline_element.get_text(strip=True)
                    news_link_element = news_card.find('a')['href']
                    news_link = news_link_element
                    news_date = extract_date(news_link)
                    
                    data.append({
                        'Heading': heading,
                        'Date': news_date,
                        'News Link': news_link
                    })

cnbc_df = pd.DataFrame(data)
cnbc_df.head(20)
```

Paper title

```python
html_text_kea = requests.get('https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/').text

soup_kea = BeautifulSoup(html_text_kea, "html.parser")

def extract_author(input_str):
    match = re.sub(r'\|', ',', input_str)
    if match:
        return match
    return None

def extract_year(input_str):
    match = re.search(r'\b\d{4}\b', str(input_str))
    if match:
        return match.group()
    return None

paper_cards=soup_kea.find_all("div",class_="panel")
data = []
    
for paper_card in paper_cards:
    paper_title_element=paper_card.find('a')
    publish_year_element=paper_card.find("p", class_="article-date")
    author_element=paper_card.find("p", class_="article-authors")
  
    title = paper_title_element.get_text(strip=True) if paper_title_element else None
    publish_year_str = publish_year_element if publish_year_element else None
    publish_year=extract_year(publish_year_str) if publish_year_str else None
    author_str = author_element.get_text(strip=True) if author_element else None
    author=extract_author(author_str) if author_str else None

    data.append({
        'Paper Title': title,
        'Paublication Year': publish_year,
        'Authors': author
    })


kea_df = pd.DataFrame(data)

kea_df.head(10)
```
